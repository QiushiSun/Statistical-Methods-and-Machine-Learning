```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVC
# Support Vector Classification.
# load data
X = [[1, 2], [2, 3], [3, 3], [2, 1], [3, 2]]
y = [1, 1, 1, -1, -1]
# train my svm classifier
clf = SVC(kernel='linear', C=1024)
clf.fit(X, y)
print("weight =", clf.coef_)
print("bias =", clf.intercept_)
print("Support Vec =", clf.support_vectors_)
```

```python
color_seq = ['coral' if v==1 else 'blue' for v in y]
plt.scatter([i[0] for i in X], [i[1] for i in X], c=color_seq)
xaxis = np.linspace(0, 5.5) # 均分x轴
slope = clf.coef_[0] # slope
print('weight:',slope)
a = -slope[0] / slope[1]
y_sep = a * xaxis - (clf.intercept_[0]) / slope[1] # 分离超平面
b = clf.support_vectors_[0]
yy_down = a * xaxis + (b[1] - a * b[0]) # 下界
b = clf.support_vectors_[-1]
yy_up = a * xaxis + (b[1] - a * b[0]) # 上界
plt.plot(xaxis, y_sep, 'k-')
plt.plot(xaxis, yy_down, 'k--')
plt.plot(xaxis, yy_up, 'k--')
plt.title('Support Vector Machine Classification')
plt.xlabel('$x^{(1)}$')
plt.ylabel('$x^{(2)}$')
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=200, facecolors='none', edgecolors='y')
plt.show()
```
